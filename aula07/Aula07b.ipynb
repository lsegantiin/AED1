```python
n = int(input("Digite a quantidade de pessoas: "))
soma_peso = 0
soma_altura = 0
maior_imc = 0
menor_imc = 1000

x = 1
while x <= n:
    peso = float(input("Digite o peso em Kg: "))
    altura = float(input("Digite a altura em M: "))
    imc = peso / (altura ** 2)
    if imc < menor_imc:
        menor_imc = imc
    if imc > maior_imc:
        maior_imc = imc
    soma_peso += peso
    soma_altura += altura
    x += 1
media_peso = soma_peso / n
media_altura = soma_altura / n
print("Imc maior = ", maior_imc, " Imc menor = ", menor_imc)
print("Peso Médio = ", media_peso)
print("Altura Média = ", media_altura)
```
