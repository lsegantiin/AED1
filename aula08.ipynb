```python
#Jogo de Advinhanças

import random
numero_sorteado = random.randint(1, 50)

while True:
    sorteio = int(input("Escolha um número para o Sorteio entre 1 e 50: "))
    if sorteio == numero_sorteado:
        print("Parabéns, você acertou!")
        break
    elif sorteio < numero_sorteado:
        print("O número sorteado é MAIOR que o seu.")
    else:
        print("O número sorteado é MENOR que o seu.")
    

```
